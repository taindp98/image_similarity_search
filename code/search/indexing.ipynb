```python
import torch
import os
from glob import glob
from tqdm import tqdm
import numpy as np
from PIL import Image
from utils import visualize_group
```

```python
_ENCODE_PATH = '../../encode'
```

```python
list_file_enc = glob(os.path.join(_ENCODE_PATH,'*'))
list_file_enc = sorted(list_file_enc)
len(list_file_enc)
```

```python
n = 3
list_file_enc_group = [sorted(list_file_enc[k:k+n]) for k in range(0, len(list_file_enc), n)]
```

```python
list_id = []
list_path = []
list_vector = []
for group in list_file_enc_group:
    _id,path,vector = group
    _id = torch.load(_id).numpy().tolist()
    list_id += _id
    path = list(torch.load(path))
    list_path += path
    vector = torch.load(vector,map_location=torch.device('cpu'))
    list_vector.append(vector)

full_vector = torch.cat(list_vector,0)
```

```python
test_enc = torch.load('../../test/encode/test_enc.h5')
# test_enc
```

```python
list_score = []
for item in tqdm(full_vector,total=full_vector.shape[0]):
    score = torch.norm(item - test_enc)
    list_score.append(score.item())
```

```python
list_score_array = np.array(list_score)
list_top_idx = list_score_array.argsort().tolist()[:5]
```

```python
np.argmin(list_score_array)
```

```python
np.sort(list_score_array)
```

```python
list_path_idx = [list_path[idx] for idx in list_top_idx]
# list_path_idx
```
